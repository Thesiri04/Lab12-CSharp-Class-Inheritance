# Lab 12 Class Inheritance

## Access overriding base class member (Methods)

การเข้าถึงสมาชิกของ Base class ที่ถูกปิดบัง ทำได้โดยการแปลงชนิดข้อมูลไปยัง BaseClass เสียก่อน แล้วจึงเข้าถึงสมาชิกที่สืบทอดมาจาก base class นั้น 

สามารถทำได้ 2 แบบ คือ  

1. ใช้วิธีการแปลงประเภท (Type Casting) หรือ การแปลงชนิดข้อมูล (Data Type Conversion)

2. ใช้ keyword `'as'`

1. สร้าง console application project

```cmd
dotnet new console --name Lab12_Ex10
```

2.เปลี่ยน code ให้เป็นดังต่อไปนี้

```csharp
// 1. ใช้วิธีการแปลงประเภท (Type Casting) หรือ การแปลงชนิดข้อมูล (Data Type Conversion)
DerivedClass dc = new DerivedClass();
Console.WriteLine("Use Type Casting or Data Type Conversion");
((BaseClass)dc).Greeting(); 

// 2. ใช้ keyword 'as'
Console.WriteLine("------------");
Console.WriteLine("Use as keyword");
(dc as BaseClass).Greeting();

class BaseClass
{
    public string Hello = "Hello From BaseClass";
    public void Greeting()
    {
        Console.WriteLine("This is method in BaseClass");
        Console.WriteLine(Hello);
    }

}
class DerivedClass : BaseClass
{
    public string Hello = "Hello From DerivedClass";
    public void Greeting()
    {
        Console.WriteLine("This is method in DerivedClass");
        Console.WriteLine(Hello);
    }
}
```

7.อธิบายสิ่งที่พบในการทดลอง

1. การแปลงประเภท (Type Casting)
- dc เป็นอ็อบเจกต์ของ DerivedClass แต่เราแปลงมันไปเป็น BaseClass โดยใช้ Type Casting ((BaseClass)dc)
- ถึงแม้ว่าจะแปลง dc เป็น BaseClass, แต่เนื่องจาก Greeting() ใน DerivedClass ได้ทับ (shadowed) เมธอดใน BaseClass, เมธอดที่ถูกเรียกจะเป็นของ DerivedClass อยู่ดี
- ผลลัพธ์: เมธอด Greeting() ของ DerivedClass จะถูกเรียก, และจะพิมพ์ข้อความจาก DerivedClass
2. การใช้ keyword as
- as ใช้เพื่อแปลง dc ไปเป็น BaseClass ถ้าแปลงสำเร็จ (หมายถึง dc เป็นอ็อบเจกต์ที่สามารถแปลงเป็น BaseClass ได้), มันจะคืนค่าอ็อบเจกต์ที่แปลงแล้ว
- ในกรณีนี้, การแปลงสำเร็จเพราะ dc เป็นอ็อบเจกต์ของ DerivedClass, แต่เมธอด Greeting() ที่ถูกเรียกยังคงเป็นของ DerivedClass เพราะมันถูกทับ (shadowed)
- ผลลัพธ์: เมธอด Greeting() จาก DerivedClass จะถูกเรียก และพิมพ์ข้อความจาก Derive
